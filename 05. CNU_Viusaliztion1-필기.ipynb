```python
# title, label 설정

plt.figure(figsize=(18, 9))

x = np.linspace(0, 3, 101)
y = np.sqrt(x)

plt.plot(x,y)
plt.title('Graph of y=x^(0.5)', fontsize = 25)
plt.xlabel('x', fontsize = 20)
plt.ylabel('y', fontsize = 20)
plt.grid()
```

```python
# plot 스타일 조절 (색상, 선 종류 등)

x = np.linspace(0, 2*np.pi, 101)
y = np.cos(x)

# plt.plot(x,y, color = 'red')
plt.plot(x,y, color = 'r', linestyle = '-.') # r 레드, b 블루, g 그린 등 # --, -. 등의 linestyle
```

```python
plt.plot(x,y, "b--") # 이런식으로 b 파란색 -- linestyle
```

```python
# 여러개의 plot도 가능

x = np.linspace(1, 4, 101)
y1 = 1/x
y2 = 1/x**2

plt.plot(x, y1,"r--", x, y2, "b-.")
plt.legend(['1/x', '1/x^2'])
```

```python
# 보통 이렇게 한다.

x = np.linspace(1, 4, 101)
y1 = 1/x
y2 = 1/x**2

plt.plot(x, y1, 'r--', label = '1/x')
plt.plot(x, y2, 'b-.', label = '1/x^2')
plt.legend(loc = 'best')
```

```python
# 연습문제 1

plt.figure(figsize = (12, 12))

x = np.linspace(0, 1, 101)
y1 = np.sqrt(x)
y2 = x ** 2
y3 = x

plt.plot(x, y1, 'b--', label = 'y = sqrt(x)')
plt.plot(x, y2, 'y.', label = 'y = x**2')
plt.plot(x, y3, 'g-.', label = 'y = x')

plt.title('Graph', fontsize = 25)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend(loc = 'best')
```

```python
# seaborn : 통계적 그래프를 잘 보여주는 라이브러리
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# 데이터를 받으면 탐색을 시작
tips.describe()
```

```python
# 결측치의 여부
tips.isna().sum()
```

```python
# 상관관계 시각화 하기

# correlation matrix 구하기
tips.corr()
```

```python
# heatmap으로 시각화하기
plt.figure(figsize=(12, 12))
sns.heatmap(tips.corr(), annot = True, annot_kws = {'fontsize':19})
plt.show()
```

```python
# scatter plot : 산점도

# relplot 으로 시각화.
# xlabel, ylabel 입력하지 않아도 자동으로 뜬다.

# 총 가격이 높을수록 받는 팁도 많을까?
# => total bill과 tip은 어느정도로 상관관계가 있을까?

sns.set(style = 'darkgrid')
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# 점심, 저녁간 팁의 차이가 있을까?
# => hue를 이용하여 데이터를 구분하기
sns.relplot(x = 'total_bill', y = 'tip',
            hue = 'time', height = 7,
            data = tips)
```

```python
# style로 데이터 구분하기
# 점심과 저녁간의 팁 중 성별에 따른 팁의 차이가 있을까?

sns.relplot(x = 'total_bill', y = 'tip',
            hue = 'time', style = 'sex', height = 5,
            data = tips) # 어질,,
```

```python
# 데이터를 다른 기준에 따라 각각 볼 수 있다.
# => col로 나누기

sns.relplot(x = 'total_bill', y = 'tip',
            hue = 'time', col = 'sex', height = 5,
            data = tips)
```

```python
# 연습문제 2

titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.describe()
```

```python
# titanic 데이터를 운임요금(fare)에 따른 연령(age)를 아래 조건 하에 relplot을 이용하여 산점도로 나타내고 분석
# 1. 생사여부(survived)를 각각 구분
# 2. 성별(sex)에 따라 산점도를 표시
sns.relplot(x = 'fare', y = 'age',
            hue = 'sex', col = 'survived', height = 5,
            data = titanic)
```

```python
# 운임요금과 연령간의 상관관계는 전혀 없다.
# 여자가 남자보다 월등히 많이 생존하였다.
```

```python
# Boxplot : 다섯 수치를 요약해주는 그래프
# 최솟값 
# 제 1사분위(Q1) : 데이터의 25% 위치
# 제 2사분위(Q2) = 중앙값(median) : 50% 위치 
# 제 3사분위(Q3) : 데이터의 75% 위치 
# 최댓값
```

```python
sns.set_theme(style='white')
sns.boxplot(x = 'total_bill', data = tips)
plt.title('Total bill boxplot')
```

```python
# boxplot 여러개 그리기
# 요일별 total_bill에 대한 boxplot

sns.boxplot(x = 'day', y = 'total_bill', data = tips)
```

```python
# 평일보다는 주말에 매출이 더 높은것을 알 수 있음
```

```python
# 요일별 팁의 박스그래프

sns.boxplot(x = 'day', y = 'tip', data = tips)
```

```python
# 토요일에 근무할경우 생각지도 않은 높은 팁을 받을수도 있다.
```

```python
# 성별을 기준으로 팁에 대한 박스그래프 그리기

sns.boxplot(x = 'sex', y = 'tip', data = tips)
```

```python
# 보통 남자가 평균보다 예상치 못한 많은 팁을 주는 경우가 있다.
```

```python
# hue 넣기

sns.boxplot(x = 'day', y = 'tip', hue = 'sex', data = tips)
```

```python
# 연습문제 3

plt.figure(figsize = (12,12))
sns.boxplot(x = 'class', y = 'age', hue = 'sex', data = titanic)
```

```python
# 등급이 높을수록 나이가 많다.
# 같은 등급 내에서도 남자보다 여자가 어린편
```

```python
# boxenplot

plt.figure(figsize = (12,12))
sns.boxenplot(x = 'sex', y = 'tip', hue = 'smoker', data = tips)
```

```python
# swarmplot

sns.swarmplot(x = 'day', y = 'tip', hue = 'smoker', 
              alpha = 0.7, data = tips)
```

```python
# violinplot

sns.violinplot(x = 'day', y = 'tip', hue = 'smoker', 
               bw = 0.25, split = True, data = tips)
```

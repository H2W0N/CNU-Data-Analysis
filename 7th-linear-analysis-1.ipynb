```python
# 임의의 데이터 100개를 뽑을 index 생성

select = np.random.randint(1, data.shape[0], 100)
select
```

```python
# select에 입력된 인덱스를 기준으로 데이터 추출하기

height_100 = data['키'][select]
weight_100 = data['몸무게'][select]

plt.scatter(weight_100, height_100)
plt.title('Korean teen health data')
plt.xlabel('weight')
plt.ylabel('height')
plt.show()
```

```python
# 문제 1

# (기울기, y절편, 데이터(주어진 값))
def line(a, b, x) :
  return a*x + b

line(1, 105, data['몸무게'][select])
plt.scatter(weight_100, height_100)
plt.plot(data['몸무게'][select], line(1, 105, data['몸무게'][select]))
```

```python
a = 1
b = 105
x = np.linspace(0, 100, 1001)
y = line(a, b, x)

plt.scatter(weight_100, height_100)
plt.plot(x, y, 'r--')
plt.text(70, 130, 'a : {}, b : {}'.format(a, b))
plt.title('Korean teen height data', fontsize=20)
plt.xlabel('weight')
plt.ylabel('height')

plt.xlim([0,100])
plt.ylim([100,190])
plt.show()
```

```python
# error를 구하기 위한 dataframe을 df에 생성
# dataframe끼리 합칠때는 concat을 사용한다.

df = pd.concat([weight_100, height_100], axis = 1)
df
```

```python
# index reset

df.reset_index(drop = True, inplace = True)
df
```

```python
# linear function으로 예측한 예상 키를 df에 첨가
a = 1
b = 105
df['예상 키'] = line(a, b, df['몸무게'])
df
```

```python
# 예상 키와 실제 키와의 차이를 계산하기.

diff = df['키'] - df['예상 키']
diff
```

```python
# RMSE 구하기

np.sqrt(sum(diff**2)/len(diff))
```

```python
# a와 b에 따른 RMSE 변화를 시각적으로 살펴보기

a, b = 0.99, 100

x = np.linspace(0, 100, 1001)
y = line(a, b, x)

df = pd.concat([weight_100, height_100], axis = 1)
df['예상 키'] = line(a, b, df['몸무게'])
diff = df['키'] - df['예상 키']

RMSE = np.sqrt(sum(diff**2)/len(diff))

# 시각화

plt.scatter(weight_100, height_100)
plt.plot(x, y, 'r--')

plt.text(70, 130, 'a : {}, b : {}'.format(a, b))
plt.text(71, 120, 'RMSE : {:1.3f}'.format(RMSE))

plt.xlim([0, 100])
plt.ylim([100, 190])
plt.show()
```

```python
# 가능한 a, b의 쌍을 모두 입력해보고 최소 RMSE 값을 뽑아보자

data_rmse = pd.DataFrame([a, b, RMSE]).T
data_rmse.columns = ['a', 'b', 'RMSE']
data_rmse
```

```python
a_range = np.linspace(0.9, 1.1, 3)
b_range = np.linspace(90, 110, 3)

a_list = []
b_list = []
rmse_list = []

df = pd.concat([weight_100, height_100], axis = 1)

for a in a_range :
  for b in b_range :
    df['예상 키'] = line(a, b, df['몸무게'])
    diff = df['키'] - df['예상 키']
    RMSE = np.sqrt(sum(diff**2)/len(diff))

    a_list.append(a)
    b_list.append(b)
    rmse_list.append(RMSE)

data_rmse = pd.DataFrame([a_list, b_list, rmse_list]).T
data_rmse.columns = ['a', 'b', 'RMSE']
data_rmse
```

```python
x = np.linspace(0, 100, 1001)

for k in range(data_rmse.shape[0]) :
  a, b, RMSE = data_rmse.iloc[k,:]

  y = line(a, b, x)

  plt.scatter(weight_100, height_100)
  plt.plot(x, y, 'r--')
  plt.text(70, 130, 'a : {}, b : {}'.format(a, b))
  plt.text(71, 120, 'RMSE : {:1.3f}'.format(RMSE))
  plt.title('Korean teen height data', fontsize=20)
  plt.xlabel('weight')
  plt.ylabel('height')

  plt.xlim([0,100])
  plt.ylim([100,190])
  plt.show()
  print('\n')
```
